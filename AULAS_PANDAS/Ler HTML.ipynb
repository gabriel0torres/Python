# Importando as libs

```python
import html5lib 
import lxml
import pandas as pd
```

# Atribuindo o html a váriavel

```python
url = 'http://tabela-cursos.herokuapp.com/index.html'

cursos_df = pd.read_html(url) #esta função procura pela tag <table> dentro da página HTML
cursos_df = cursos_df[0]
```

# mostrando o resultado

```python
display(cursos_df)
```

# alterando o index de cursos

```python
cursos_df.rename(columns={'Nome do curso' : 'nome_do_curso'})
```

```python
cursos_df['ID'] = cursos_df.index + 1
cursos_df = cursos_df.set_index('ID')
display(cursos_df)
```
